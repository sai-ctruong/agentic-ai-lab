```python
import os
from dotenv import load_dotenv
import google.generativeai as genai
load_dotenv()

gemini_api_key= os.getenv("GEMINI_API_KEY")
genai.configure(api_key= gemini_api_key)
print("API client successfully configured")

print(gemini_api_key[:5])
```

```python
my_message= "Write me three interesting facts about people."
print(f"Sending messages to AI: '{my_message}'")
```

```python
model = genai.GenerativeModel("gemini-2.5-flash")
response= model.generate_content(my_message)
```

```python
ai_reply_content=response.text

print("\n AI's reply: \n")
print(f"{ai_reply_content}")
```
